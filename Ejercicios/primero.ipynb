```python
# crenado una session en pyspark 

# Importar la clase principal para crear la sesión de Spark
from pyspark.sql import SparkSession
ruta = "../documentos/"


# Crear o recuperar una sesión de Spark
# appName es opcional pero se recomienda para identificar el job
spark = SparkSession.builder \
    .appName("mi_app_pyspark") \
    .master("local[3]") \
    .getOrCreate()
```

```python
sc = spark.sparkContext # con esto podemos usar el contexto de spark para crear RDDs
```

### Pregunta 2

```python
# Crear un RDD vacío
rdd_vacio = sc.emptyRDD()

# crenado un RDD vacion con 5 particiones
rdd_vacio5 = sc.parallelize([], 5)

print(rdd_vacio.collect()) # esto devuelve una lista vacia

lista = [1,2,3,5,7,11,13,17,19]

rdd = sc.parallelize(lista)

print(rdd.collect())
```

Comando para saber propiedades basicas de los Rdd

```python
print(rdd.getNumPartitions())
print(rdd.count())
```

### Pregunta 3

```python
# creando un rdd, con los nuemeros primos mayores a 10
primos_mayores_10 = rdd.filter(lambda x: x > 10)
print(primos_mayores_10.collect())
```
