```python
# Importar la clase principal para crear la sesión de Spark
from pyspark.sql import SparkSession
ruta = "../documentos/"


# Crear o recuperar una sesión de Spark
# appName es opcional pero se recomienda para identificar el job
spark = SparkSession.builder \
    .appName("mi_app_pyspark") \
    .getOrCreate()
```

```python
# Obtener SparkContext para trabajar con RDDs
sc = spark.sparkContext
```

```python
# Crear un RDD vacío
rdd_vacio = sc.emptyRDD()

# Crear un RDD vacío con 3 particiones
rdd_vacio3 = sc.parallelize([], 3)

# Ver cuántas particiones tiene el RDD
num_particiones = rdd_vacio3.getNumPartitions()
print(num_particiones)

# Crear un RDD con datos
rdd = sc.parallelize([1, 2, 3, 4, 5])
```

```python
# Crear un RDD a partir de un archivo de texto
rdd_texto = sc.textFile(ruta  + "rdd_source.txt")

# Obtener todos los elementos del RDD (acción)
lineas = rdd_texto.collect()
print(lineas)
```

```python
# Crear un RDD donde cada elemento es (ruta_archivo, contenido_completo)
rdd_texto_completo = sc.wholeTextFiles(ruta + "rdd_source.txt")

# Acción: traer los datos al driver
archivos = rdd_texto_completo.collect()

print(archivos)
```

```python
# Transformación: aplicar una función a cada elemento del RDD
rdd_suma = rdd.map(lambda x: x + 1)

# Acción: traer los resultados al driver
resultado = rdd_suma.collect()

print(resultado)
```

```python
# Crear un DataFrame a partir de datos en memoria
data = [(1, "jose"), (2, "juan")]
columnas = ["id", "nombre"]

df = spark.createDataFrame(data, columnas)

# Mostrar el contenido del DataFrame
df.show()
```

```python
# Convertir un DataFrame a RDD
rdd_df = df.rdd

# Acción: traer los elementos del RDD al driver
resultado = rdd_df.collect()

print(resultado)
```
